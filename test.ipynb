```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "meta-llama/Llama-3.2-1B"

# 1. 加载 tokenizer 和 model
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",           # 自动分配到 GPU/CPU
    torch_dtype="auto",          # 自动选择 dtype
    trust_remote_code=True       # Qwen 系列需要
)

# 2. 查看 eos token 信息
print("EOS token:", tokenizer.eos_token)
print("EOS token id:", tokenizer.eos_token_id)

# 3. 验证该 token
decoded = tokenizer.decode([tokenizer.eos_token_id])
print("Decoded EOS token:", repr(decoded))
```

```python
print("Model loaded from:", model.config.name_or_path)
print("Tokenizer loaded from:", tokenizer.name_or_path)
```

```python
import transformers

model_id = "meta-llama/Llama-3.1-8B-Instruct"

# 加载 tokenizer
tokenizer = transformers.AutoTokenizer.from_pretrained(model_id)

# 输入文本
text = "Hello, how are you?"

# 编码（得到 token ids）
input_ids = tokenizer.encode(text, add_special_tokens=True)
print(tokenizer.decode(input_ids))
```

```python
from transformers import AutoTokenizer
model_id = "meta-llama/Llama-3.2-1B-Instruct"
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

tokenizer = AutoTokenizer.from_pretrained(model_id)

# Rendered prompt
rendered_prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
print("\n=== Rendered Prompt ===")
print(rendered_prompt)
```
